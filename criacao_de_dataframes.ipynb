```python
import pandas as pd
import numpy as np

datas = pd.date_range('20190101', periods = 60, freq='D')
df = pd.DataFrame(np.random.randn(60,5), index = datas, columns=list('ABCDE'))
df.shape #mostra linhas e colunas
df.head(3) #construir as linhas
df['F'] = 1 #coluna a mais
df.head(10) #add 10 linhas
```

```python
#df['G'] = [10,3,5,7] #nao tem 60 colunas
df['G'] = range(60) #adicionar novas colunas
```

```python
df['Produto'] = df['A'] * df['B']
df.head(3)
```

```python
df['D'] = 88
df.head(3)
```
